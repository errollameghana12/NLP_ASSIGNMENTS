```python
''' 
1. Difference between "Love" and "love" in NLP
In NLP, "Love" and "love" are treated as two different words if case conversion is not applied. This can increase the number of unique words unnecessarily. To avoid this, we convert all text to lowercase so both become the same word "love".
2. What happens if stopwords are not removed?
If stopwords are not removed, the dataset will contain many unnecessary words like "is", "the", "and". This increases the size of the data and may reduce model performance because these words do not add much meaning.
3. Two real-world scenarios where removing stopwords is harmful
In sentiment analysis: Removing words like "not" changes the meaning of a sentence. Example: "I am not happy" becomes "I happy".
In question answering: Words like "what", "when", "where" are important to understand the question.
4. Difference between stemming and lemmatization
Stemming is a simple method that removes suffixes from words, but sometimes the output is not a real word. Lemmatization is more advanced and converts words into their correct base form using a dictionary. Stemming is faster but less accurate, while lemmatization is slower but more meaningful.
'''
```

```python
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()
    allowed = {"no", "not"}
    tokens = [word for word in tokens if len(word) > 2 or word in allowed]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence


text = "WOW!!! This is GREAT!!! I have 2 dogs"
tokens, cleaned = preprocess_text(text)

print("Tokens:", tokens)
print("Cleaned Sentence:", cleaned)
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)
    print("Original Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean)
    print("-"*50)
```

```python
def token_analysis(tokens):
    if not tokens:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total

    return total, unique, avg_len


for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)
    total, unique, avg = token_analysis(tokens)

    print("Sentence:", sentence)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg, 2))
    print("-"*50)
```

```python
all_tokens = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:", counter.most_common(10))
print("Top 5 Least Frequent Words:", counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


result = full_pipeline(test_sentences)

print("All Tokens:", result["tokens"])
print("Clean Sentences:", result["clean_sentences"])
```
